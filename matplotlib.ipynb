```python
import matplotlib.pyplot as plt
```

```python
#intializing the data
x=[10,20,30,40]
y=[20,30,40,50]
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y)
#adding the title
plt.title("simple plot")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[3,2,2])
plt.title("1st subplot")
plt.subplot(1,2,2)
plt.plot([2,3,3],[3,2,2],"b^")
plt.title("2nd subplot")
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
ax1.plot([1,2,3])
ax2=fig.add_subplot(2,1,1,facecolor='w')
ax2.plot([1,2,0.5])
ax3=fig.add_subplot(2,2,4)
ax3.plot([1,2,3],[1,2,3],"r^")
```
